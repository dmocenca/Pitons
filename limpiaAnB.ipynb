```python
data3=data2.copy()
```

```python
# Check the data type of 'host_response_rate'
print(data3['host_response_rate'].dtype)
```

```python
# Assuming 'host_response_rate' is an object type with percentage signs
data3['host_response_rate'] = data3['host_response_rate'].str.replace('%', '', regex=False)
data3['host_response_rate'] = pd.to_numeric(data3['host_response_rate'], errors='coerce')

# Now check the data type again
print(data3['host_response_rate'].dtype)
```

```python
#Sustituir con valores promedio o media
data3["host_response_rate"]=data3["host_response_rate"].fillna(round(data3["host_response_rate"].mean(),1))
```

```python
# Assuming 'host_response_rate' is an object type with percentage signs
data3['host_acceptance_rate'] = data3['host_acceptance_rate'].str.replace('%', '', regex=False)
data3['host_acceptance_rate'] = pd.to_numeric(data3['host_acceptance_rate'], errors='coerce')

# Now check the data type again
print(data3['host_acceptance_rate'].dtype)
```

```python
#Sustituir con valores promedio o media
data3["host_acceptance_rate"]=data3["host_acceptance_rate"].fillna(round(data3["host_acceptance_rate"].mean(),1))
```
